```python
from bs4 import BeautifulSoup
import seaborn as sns
import pandas as pd
import requests
import openpyxl
```

```python
base_url = 'https://www.booking.com/searchresults.html?ss=New+York%2C+New+York+State%2C+United+States&ssne=India&ssne_untouched=India&label=gen173nr-1FCAEoggI46AdIM1gEaGyIAQGYATG4ARfIAQzYAQHoAQH4AQKIAgGoAgO4Av2P3rAGwAIB0gIkM2Y5ZWQ0ZTYtYTE1ZC00OGVmLThiYmMtMmYyZmU5NjkyZDkw2AIF4AIB&aid=304142&lang=en-us&sb=1&src_elem=sb&src=index&dest_id=20088325&dest_type=city&ac_position=1&ac_click_type=b&ac_langcode=en&ac_suggestion_list_length=5&search_selected=true&search_pageview_id=005c2ffebd5200d1&ac_meta=GhAwMDVjMmZmZWJkNTIwMGQxIAEoATICZW46A25ld0AASgBQAA%3D%3D&checkin=2024-04-18&checkout=2024-04-24&group_adults=2&no_rooms=1&group_children=0'
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}
```

```python
hotel_results = []
```

```python
for page_number in range(0, 250,25):
    url = base_url.format(page_number)

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    for el in soup.find_all("div", {"data-testid": "property-card"}):
        hotel_results.append({
            'name':el.find("div", {"data-testid": "title"}).text.strip(),
            'description': el.find('div', {'data-testid': 'recommended-units', 'class': 'ccdd44706b'}).text.strip(),
            "location": el.find("span", {"data-testid": "address"}).text.strip(),
            "pricing": el.find("span", {"data-testid": "price-and-discounted-price"}).text.strip(),
            "rating": el.find("div", {"data-testid": "review-score"}).text.strip().split(" ")[0],
        })

# Create DataFrame from the collected results
hotels = pd.DataFrame(hotel_results)
```

```python
hotels.to_excel('datp.xlsx',index=False)
```
