```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import kagglehub
```

```python
path = kagglehub.dataset_download("balraj98/berkeley-segmentation-dataset-500-bsds500")
print("Path to dataset files:", path)
```

```python
dataset_path = os.path.join(path, 'images/test')
```

```python
def apply_canny(image_path):

    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Image not found.")
        return None

    return image, cv2.Canny(image, 100, 200)
```

```python
def show_images(original, edges):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edge Detection")
    plt.axis('off')

    plt.show()
```

```python
image_files = os.listdir(dataset_path)

image_path = os.path.join(dataset_path, image_files[0])

original_image, edges_image = apply_canny(image_path)

if original_image is not None and edges_image is not None:
    show_images(original_image, edges_image)
```
